```python
def detectar_outliers(df, col):
  Q1 = df[col].quantile(0.25)
  Q3 = df[col].quantile(0.75)
  IQR = Q3 - Q1
  outliers = df[(df[col] < Q1 - 1.5 * IQR) | (df[col] > Q3 + 1.5 * IQR)]
  return outliers
```

```python
outliers_estatura = detectar_outliers(df, "talla_de_zapato_(cm)")
print('Outliers: ')
print(outliers_estatura)
```

```python
columnas = ['edad', 'peso_(kg)', 'estatura_(cm)', 'talla_de_zapato_(cm)']
for col in columnas:
  outliers = detectar_outliers(df, col)
  print(f'Outliers for {col}: ')
  print(outliers)
  print('-' * 30)
```

```python
for col in columnas_numericas:
  estadisticas = stats.describe(df[col].dropna())
  print(f'Estadísticas descriptivas para {col}:')
  print(estadisticas)
```

```python
# Método 1: Test de Normalidad (Shapiro-Wilk)

for col in columnas_numericas:
  statistic, p_value = stats.shapiro(df[col].dropna())
  print(f'\nTest de Normalidad (Shapiro-Wilk) para {col}:')
  print(f'Estadístico: {statistic:.4f}, P-valor: {p_value:.4f}')
  if p_value > 0.05:
    print('Conclusión: No se rechaza la hipótesis nula, los datos pueden provenir de una distribución normal.')
  else:
    print('Conclusión: Se rechaza la hipótesis nula, los datos no provienen de una distribución normal.')

# Método 2: Coeficiente de Asimetría (Skewness)

for col in columnas_numericas:
  skewness_value = stats.skew(df[col].dropna())
  print(f'\nCoeficiente de Asimetría (Skewness) para {col}:')
  print(f'Valor: {skewness_value:.4f}')
  if skewness_value == 0:
    print('Conclusión: La distribución es simétrica.')
  elif skewness_value > 0:
    print('Conclusión: La distribución tiene una cola derecha más larga (asimétrica positiva).')
  else:
    print('Conclusión: La distribución tiene una cola izquierda más larga (asimétrica negativa).')

# Método 3: Test T de Student para una muestra

media_global = df[columnas_numericas].mean().mean()
print(f'\nMedia global de las columnas numéricas: {media_global:.4f}')

for col in columnas_numericas:
  # Vamos a usar la media global calculada como el valor hipotético
  popmean = media_global
  statistic, p_value = stats.ttest_1samp(df[col].dropna(), popmean=popmean)
  print(f'\nTest T de Student para una muestra en {col} (comparado con la media global {popmean:.4f}):')
  print(f'Estadístico T: {statistic:.4f}, P-valor: {p_value:.4f}')
  if p_value > 0.05:
    print('Conclusión: No se rechaza la hipótesis nula, la media de la columna no difiere significativamente de la media global.')
  else:
    print('Conclusión: Se rechaza la hipótesis nula, la media de la columna difiere significativamente de la media global.')
```
